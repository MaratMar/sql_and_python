```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
df = pd.read_csv(r'.\Lec1\lec1\housing.csv')
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('MEDV', axis = 1), 
                                                    df.MEDV, 
                                                    test_size = 0.2, 
                                                    random_state = 1904)
```

```python
X_train
```

```python
X_test
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
train_pred_medv = model.predict(X_train) 
test_pred_medv = model.predict(X_test)
```

```python
train_pred_medv
```

```python
test_pred_medv
```

```python
print ('Error MSE Train =', sqrt(mean_squared_error(y_train, train_pred_medv)))

print ('Error MSE Test =', sqrt(mean_squared_error(y_test, test_pred_medv)))
```
